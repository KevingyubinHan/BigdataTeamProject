# 층화추출

### seg1

```python
need_cols = ['BAD','Grd_NK0200_000','Grd_RK0400_000','Grd_RK0400_700',]
```

```python
seg1_grp = seg1_dev_df.groupby(by = need_cols, as_index = False).agg(
{
    'no': ('count','unique')
})
seg1_grp.columns = need_cols + ['cnt','no']
```

```python
seg1_grp
```

```python
def _sampling(no, sample_pct):
    N = len(no)
    vv = np.random.choice(no, replace = False, size = int(N * sample_pct))
    return vv
```

```python
def _drop(total, selected):
    rest = np.setdiff1d(total, selected)
    return rest
```

```python
np.random.seed(111)
seg1_grp['tr_no'] = seg1_grp['no'].apply(lambda x: _sampling(x, sample_pct = 0.6),)
```

```python
seg1_grp['val_tst_no'] = seg1_grp[['no','tr_no']].apply(lambda x: _drop(x['no'], x['tr_no']), axis = 1)
```

```python
np.random.seed(111)
seg1_grp['val_no'] = seg1_grp['val_tst_no'].apply(lambda x: _sampling(x, sample_pct = 0.5))
```

```python
seg1_grp['tst_no'] = seg1_grp[['val_tst_no', 'val_no']].apply(lambda x: _drop(x['val_tst_no'], x['val_no']), axis = 1)
```

```python
seg1_grp
```

### 해당 조합에 수가 적은 경우 train/valid/test 순으로 넣어주기

```python
# 대상이 1건인 경우 tst에 들어감
## tst에 들어가는 대상 trn으로 들어가게 바꾸기

# 대상이 2건인 경우 trn, tst에 들어감
## tst에들어가는 대상 val으로 들어가게 바꾸기

def unique_one_to_train(tr_no,val_no, tst_no):
    if len(tr_no) == 0:
        a = tr_no.copy()
        b = tst_no.copy()
        
        tr_no = b
        tst_no = a
        
        return tr_no, val_no, tst_no
    
#     elif len(val_no) < len(tst_no):
#         a = val_no.copy()
#         b = tst_no.copy()
        
#         val_no = b
#         tst_no = a
        
#         return tr_no, val_no, tst_no
    else:
        return tr_no, val_no, tst_no
```

```python
in_out_cols = ['tr_no','val_no','tst_no']
```

```python
seg1_grp[in_out_cols] = seg1_grp[in_out_cols].apply(lambda x: unique_one_to_train(x['tr_no'], x['val_no'],  x['tst_no']),
                                                                          axis = 1, result_type = 'expand')
```
