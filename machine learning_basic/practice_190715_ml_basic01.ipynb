## 행 선택

### boolean filter 생성 >> 적용

````
loan_df2.sort_values(by=['term'], axis=0, inplace=True, ascending=True)
````

````
# Boolean Filter 생성
'''
특정 컬럼(들)의 참/거짓 연산을 통해서 산출된 True, False로 이루어진 Series
'''


bool_filter_example_se = loan_df2['term'] == ' 60 months'
````

````
# 처음 5개 확인
bool_filter_example_se[:5]
````

````
# 마지막 5개 확인
bool_filter_example_se[-5:]
````

````
# 필터를 적용하여 60개월 데이터만 추출하기
loan_60mon_df = loan_df2[bool_filter_example_se].copy()
print('필터적용전 shape: ', loan_df2.shape)
print('필터적용후 shape: ', loan_60mon_df.shape)
````

````
loan_60mon_df.head()
````

### boolean filter 생성 & 적용

````
loan_60mon_df = loan_df2[loan_df2['term']== ' 60 months'].copy()
print('필터적용후 shape: ', loan_60mon_df.shape)
````

### 복수의 boolean fiter 생성 >> 적용

````
# 만기 60개월 이상 필터
bf_term = loan_df2['term'] == ' 60 months'

# 이자율 20% 이상 필터
bf_int_rate = loan_df2['int_rate'] >= 20.0

# 만기 60개월이면서 이자율 20%이상 필터
bf_term_int_rate = bf_term & bf_int_rate

loan_60mon_high_df = loan_df2[bf_term_int_rate].copy()
loan_60mon_high_df.head()
````

### 복수의 boolean fiter 생성 & 적용
[비트연산자](https://wikidocs.net/1161)

````
# 만기 60개월이면서(AND) 이자율 20%이상 필터 (& 연산자)
loan_60mon_high_df = loan_df2[(loan_df2['term'] == ' 60 months') & (loan_df2['int_rate'] >= 20.0)].copy()
loan_60mon_high_df.head()
````

````
# 만기 60개월이거나(OR) 이자율 20%이상 필터 (| 연산자)
loan_60mon_or_high_df = loan_df2[(loan_df2['term'] == ' 60 months') | (loan_df2['int_rate'] >= 20.0)].copy()
loan_60mon_or_high_df.head()
````

### boolean filter의 여집합

````
# 만기 60개월이 아닌 필터 (| 연산자)
loan_not_60mon_df1 = loan_df2[loan_df2['term'] != ' 60 months'].copy()
loan_not_60mon_df2 = loan_df2[~(loan_df2['term'] == ' 60 months')].copy()

print(loan_not_60mon_df1.shape)
print(loan_not_60mon_df2.shape)
````

## 중복 행 제거

### 중복 여부 boolean fiter 생성 >> 적용
[duplicted](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.duplicated.html?highlight=duplicated#pandas.DataFrame.duplicated)

````
# 값 정렬
loan_df2.sort_values(by=['loan_amt', 'funded_amt', 'term', 'int_rate',], axis=0, inplace=True, ascending=True)
loan_df2.tail(8)
````

````
nodup_cols = ['loan_amt', 'funded_amt', 'term', 'int_rate',]
bf_dup = loan_df2.duplicated(subset=nodup_cols, keep='first')
bf_dup[-8:]
````

````
# boolean fiter 적용
loan_nodup_df = loan_df2[~bf_dup].copy()
loan_nodup_df.tail(5)
````

### 중복 제거
[drop_duplicates](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html?highlight=drop_duplicates#pandas.DataFrame.drop_duplicates)

````
nodup_cols = ['loan_amt', 'funded_amt', 'term', 'int_rate',]
loan_nodup_df = loan_df2.drop_duplicates(subset=nodup_cols, keep='first', inplace=False)
loan_nodup_df.tail(5)
````
